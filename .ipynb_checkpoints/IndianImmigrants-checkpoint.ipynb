#Main Code

```python
import pandas as pd#Importing Necessary Packages
import numpy as np
from matplotlib import animation
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
import geoplot
import warnings
```

```python
immigrants = pd.read_csv(r"Immigrantsdata.csv",low_memory=False)
immigrants.info()#check the info of main dataset
```

```python
immigrants
```

```python
ind_img = immigrants.loc[immigrants.loc[:,"CO2"] == "IND",:]#filtering only immegrents who have originated in India
ind_img
```

```python
ind_img_coutemp = ind_img.loc[ind_img["Variable"] == "Inflows of foreign population by nationality",["COU","Value"]]
ind_img_cou = pd.DataFrame(ind_img_coutemp.groupby(pd.Grouper("COU")).sum())#dataframe is created that comprise of country code and sum of immigrants in each country 
ind_img_cou
```

# MAP PLOT

```python
x = pd.DataFrame(list(ind_img_cou.index),columns = ["code"],index = list(ind_img_cou.index))
len(x)#no of countrys that Indian Immegrate to
```

```python
ind_img_cou1 = pd.concat([ind_img_cou,x],axis = 1,ignore_index=True)
ind_img_cou1.columns = ["Values","Codes"]
```

```python
world = gpd.read_file(
    gpd.datasets.get_path('naturalearth_lowres')
)
world#importing data set containing map geometry for each country
```

```python
ind_img_cou1.reset_index(drop = True,inplace=True)
ind_img_cou1
```

```python
world_codes = pd.concat([world.iso_a3,pd.Series(world.geometry)],axis = 1)
world_codes["Codes"] = world_codes["iso_a3"]
world_codes.drop("iso_a3",axis= 1,inplace = True)
world_codes
```

```python
world_code1 = world_codes.sort_values("Codes").reset_index(drop=True)
world_code1
```

```python
plt_data = world_code1.combine_first(ind_img_cou1)
plt_data.info()
```

```python
world_code1.loc[world_code1["Codes"] == "AUS","Values"] = ind_img_cou1.loc[ind_img_cou1["Codes"] == "AUS","Values"]
world_code1.loc[world_code1["Codes"] == "AUS","Values"]
data = ['GT' + '0'*(3-len(str(i))) + str(i) for i in range(0, len(world_code1.Codes))]
world_code1['Values'] = data
str(float(ind_img_cou1.loc[ind_img_cou1["Codes"] == "AUS","Values"]))
num = str(float(ind_img_cou1.loc[ind_img_cou1["Codes"] == "AUS","Values"]))
world_code1.loc[world_code1["Codes"] == "AUS","Values"]
world_code1["Values"].replace(list(world_code1.loc[world_code1["Codes"] == "AUS","Values"])[0],str(float(ind_img_cou1.loc[ind_img_cou1["Codes"] == "AUS","Values"])),inplace=True)
```

```python
world_code1.loc[world_code1["Codes"] == "AUS","Values"]
```

```python
counter = 0
for i in world_code1["Codes"]:
    for j in ind_img_cou1["Codes"]:
        if i == j:
            print(type(j))
            world_code1["Values"].replace(list(world_code1.loc[world_code1["Codes"] == i,"Values"])[0],str(float(ind_img_cou1.loc[ind_img_cou1["Codes"] == i,"Values"])),inplace=True)
            print(world_code1.loc[world_code1["Codes"] == i,"Values"])
            counter = counter + 1
            break
```

```python
print(world_code1)
```

```python
for i in world_code1["Values"]:
    if i.startswith("GT"):
        world_code1["Values"].replace(i,np.nan,inplace=True)
```

```python
world_code2 = gpd.GeoDataFrame(world_code1)
```

```python
world_code2["Values"] = pd.to_numeric(world_code2["Values"])
```

```python
geoplot.choropleth(world_code2["geometry"], hue = (world_code2["Values"]/1000),figsize=(40,20),legend=True)
plt.title("Spread of Indian Immigrants across the world")
warnings.filterwarnings("ignore")
plt.show()
```

```python
e = ind_img_cou1.nlargest(10,"Values")
col=['#FCD1D1','#ECE2E1','#D3E0DC','#AEE1E1','#BAD7DF','#FFE2E2','#99DDCC','#30E3CA','#11999E','#40514E']
plt.figure(figsize=(40,10)).gca().bar(e["Codes"],e["Values"]/1000, color=col)
plt.title("Countries With The Largest Share Of Indian Immigrants Top 10")
plt.xlabel("Country codes")
plt.ylabel("no of Immigrants in 1000")
plt.show()
```

#Donut chart and Pie Chart for largest and smallest number of Indian Immigrants 

```python
fig1, ax1 = plt.subplots()
b=ind_img_cou1.nlargest(5,"Values")
explode = (0, 0, 0, 0,0)
col=['#6C4AB6','#8D72E1','#8D9EFF','#B9E0FF','#9ED5C5']
f=plt.pie(b['Values'],explode=explode ,labels=b['Codes'], autopct='%1.1f%%',shadow=True, startangle=90,wedgeprops = { 'linewidth' : 7, 'edgecolor' : 'white' },colors=col)
my_circle=plt.Circle( (0,0), 0.7, color='white')
plt.gcf().set_size_inches(8, 8)
plt.gca().add_artist(my_circle)
plt.title("Countries With The Largest Share Of Indian Immigrants")
plt.show()
```

```python
fig1, ax1 = plt.subplots()
b=ind_img_cou1.nsmallest(5,"Values")
col=['#65647C','#8B7E74','#C7BCA1','#F1D3B3','#F0CAA3']
plt.pie(b['Values'],labels=b['Codes'], autopct='%1.1f%%',shadow=True, startangle=90,colors=col)
plt.gcf().set_size_inches(8, 8)
plt.title("Countries With The Smallest Share Of Indian Immigrants")
plt.show()
```
